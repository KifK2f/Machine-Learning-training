```python
# Importer le librairies nécessaires 
import numpy as np #Pour tout ce qui est mathématiques 
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression # Car on va utiliser une regression linéaire
```

```python
#Emplacemnet du fichier courant
import os
os.getcwd()
```

```python
#Importer le dataset
Data = pd.read_excel('datas/VoitureOccasion.xlsx')

X = Data.iloc[:, :-1]
Y = Data.iloc[:, -1]
```

```python
#Gérer les variables catégoriques
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer # Permet d'appliquer des transformations sur certaines colonnes spécifiques

# 3 c'est l'indice de la colone Marque sachant que la première colonne a poiyr indice 0
# remainder='passthrough' # Garde les autres colonnes sans modification
ct = ColumnTransformer([("Marque", OneHotEncoder(), [3])], remainder='passthrough' )
X = ct.fit_transform(X)
```

```python
X
```

```python
#Diviser le dataset
X_train, X_test, Y_train, Y_test =  train_test_split(X, Y, test_size=0.2)
```

```python
X_train
```

```python
#Construire le modèle
regressor = LinearRegression()
regressor.fit(X_train, Y_train)
```

```python
#Faire des prédictions
Y_pred = regressor.predict(X_test)
```
