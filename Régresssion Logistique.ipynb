```python
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
#Imprter le dataset
dataset = pd.read_csv('datas/diabetes.csv')
```

```python
#Vuisualisation des données
dataset.head()
```

```python
#Afficher un graphique qui montre qu'il y'a un lien fort entre l'Outcome(est diabétiaque ou non et est la variable cible) et une autre variabme
plt.scatter(dataset.Age, dataset.Outcome)# Trouver s'il y'a un lien entre l'age et l'OutCome
```

```python
#Suppresson d'une variable qui n'a rien avoir avec notre prédiction
# dataset.drop(['NomColonneASuppr'], axis='columns' , inplace=True)
# axis='columns'  pour dire qu'on va supprimer une colonne (variable à pas une ligne
# inplace=True car au lieu d'écrire dataset = dataset.drop(), pour dire que les modifications se feront sur ce dataset
```

```python
X = dataset.iloc[:, :-1].values #Toutes les colonnes sauf la dernière
y = dataset.iloc[:, -1].values # La dernière colonne
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
#Construction du modèle
classifier = LogisticRegression(solver='liblinear', random_state=0)
classifier.fit(X_train, y_train)
```

```python
#Faire de nouvelles prédictions
y_pred = classifier.predict(X_test) #les résultats prédis par notre nod_le seront dans cette variable

# Demandeer au modèle de renvoyer un score permettat de un calcul entre les X_test et les y_test avec le valeurs calculés en utilisant le classifier
classifier.score(X_test, y_test)
```
