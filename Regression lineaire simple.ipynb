```python
# Importr les librairies
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
#Importer le datest
Data = pd.read_csv('datas/LoyerMaisons.csv', sep=";") #Car ici le séparateur est différent de ,
```

```python
Data
```

```python
print(f"-------Forme du dataset -------\n {Data.shape}")
print(f"-------Voir les valeus manquantes ------- \n {Data.isnull().sum()}")
print(f"-------Les 5 premières lignes ------- \n {Data.head()}")
print(f"-------Les 5 dernières lignes ------- \n {Data.tail()}")
print(f"-------La taille totale du dataset ------- \n {len(Data)}")

print(f"-------Lignes contenant des valeurs manquantes (NaN) -------\n {Data[Data.isnull().any(axis=1)]}")
```

```python
# Afficher nuage de points
plt.scatter(Data['surface'], Data['loyer']) #Ca a 2 paramètres en abscisses on a Data['surface'], et en ordonnée Data['loyer']
# Nb : Ici l'axe des ordonnées est en x1e7 = x10 000 000
```

```python
# Exclure(ignorer) les données dont le loyer supérieur à 1 000 000 FCFA
Data = Data[Data['loyer'] <= 1000000] 
# Signe <= car on ne veut garder que ceux qui sont inférieur ou égal à  1 000 000 FCFA = # Exclure(ignorer) les données dont le loyer supérieur à  1 000 000 FCFA
```

```python
# Afficher nuage de points
plt.scatter(Data['surface'], Data['loyer'], color='red') #Ca a 2 paramètres en abscisses on a Data['surface'], et en ordonnée Data['loyer']
# Nb : Ici l'axe des ordonnées est en x1e6 = x1 000 000
```

```python
X = Data.iloc[:, :-1].values #Toutes les colonnes sauf la dernière
Y = Data.iloc[:, -1].values # La dernière colonne
```

```python
#Diviser le dataset 
# Donc on aura des variables contenant des sous ensembles du Dataframe Data
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.33)
# 1/3 = 33% = environ 0.33 des données seront utilisées pour le test, le reste (67%) environ 0.67 pour l'entraînement
```

```python
#La taille de chaque variable
# Nb : Ona conservé que les offres <= 1 000 000 FCFA donc c'est normal que le dataset ne contiennet pas les 239 lignes
# Nb : .0% pour afficher sous forme de pourcentage
print(f"Taille de X = {len(X)}, celui de X_train est {len(X_train)} et celui de X_test est {len(X_test)}")
print(f"Proportion de X_train = {len(X_train)/len(X):.0%}")
print(f"Proportion de X_test = {len(X_test)/len(X):.0%}")
print("-------------------")
print(f"Taille de Y = {len(Y)}, celui de Y_train est {len(Y_train)} et celui de Y_test est {len(Y_test)}")
print(f"Proportion de Y_train = {len(Y_train)/len(Y):.0%}")
print(f"Proportion de Y_test = {len(Y_test)/len(Y):.0%}")
```

```python
# Vérification des Dimensions (Nombre de coordonnées nécessaires)
# 2D car c'est un tableau : il faut 2 coordonnées (Ligne, Colonne) pour isoler une donnée
# 1D car c'est un vecteur : 1 seule coordonnée (la position) suffit pour trouver la cible.
print(f"Dimension de Data (Dataframe complet) : {Data.ndim}D")
print(f"Dimension de X (Matrice variables explicatives)  : {X.ndim}D")
print(f"Dimension de Y (Vecteur cible)      : {Y.ndim}D")
print(f"Dimension de X_train (Sous-matrice) : {X_train.ndim}D")
print(f"Dimension de X_test (Sous-matrice) : {X_test.ndim}D")
print(f"Dimension de Y_train (Sous-matrice) : {Y_train.ndim}D")
print(f"Dimension de Y_test (Sous-matrice) : {Y_test.ndim}D")

# Vérification des Formes (Nombre de lignes et de colonnes)
print(f"Forme de Data    : {Data.shape}")
print(f"Forme de X       : {X.shape}")
print(f"Forme de X_train : {X_train.shape}")
print(f"Forme de X_test : {X_test.shape}")
print(f"Forme de Y       : {Y.shape}")
print(f"Forme de Y_train : {Y_train.shape}")
print(f"Forme de Y_test : {Y_test.shape}")
```

```python
X
```

```python
#Construire le modèle

#Créer le modèle
# On instancie la classe LinearRegression
# On utilise le nom 'regressor' car c'est une convention pour un modèle de régression
regressor = LinearRegression()

# Entrainer le modèle en utilisant les données d'entrainements
# La méthode .fit() permet au modèle d'apprendre la relation (la droite) 
# entre le(s) caractéristique(s) (X_train) et le(s) cible(s) (Y_train)
regressor.fit(X_train, Y_train)
```

```python
# Faire dees prédictions 
# Prédire les prix de loyer des maisons qui se trouvent dans la base de test 
# qu'on va comparrer apres avec Y_Test
Y_pred = regressor.predict(X_test) #Cette variable repésente les prix prédis par notre modèle
```

```python
# Afficher les prédiction concernant les maisons qui se trouvent dans la base de test (X_test)
Y_pred
```

```python
# Comparer Y_pred avec Y_Test
print(Y_test)
print("----------")
# Prédiction pour une donnée qui n'existe mm pas dans la base (Ni dans train et ni dans test)
# Par exempel une maison ayant une surface de 125 et on va prédire son loyé
print(f"Pour une maison de surface 125 , le prix est : {regressor.predict( [ [125] ] )}")
```

```python
#Visualisr les résultats
#Nuage de points 
plt.scatter(X_test, Y_test, color='red')

# Droite de regresion et prédire les sorties concernant X_train
plt.plot(X_train, regressor.predict(X_train), color='blue')

#Ajouter un titre
plt.title("Evolution des loyers par surface")
plt.xlabel("Surface")
plt.ylabel("Loyer")
```
